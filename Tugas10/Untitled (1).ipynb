```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_excel('Dataset.xlsx')
dataset.keys()
```

```python
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
X = np.asarray(dataset)
print(X)
```

```python
plt.scatter(X[:,1], X[:,4], label='True Position')
plt.xlabel("Tanggal")
plt.ylabel("Penjualan")
plt.title("Grafik Penyebaran")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X[:,1], X[:,4], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("Tanggal")
plt.ylabel("penjualan")
plt.title("Grafik Hasil Klasterisasi")
plt.show()
```

```python
plt.scatter(X[:,1], X[:,4], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,4], color='black')
plt.xlabel("Tanggal")
plt.ylabel("Penjualan")
plt.title("Grafik Hasil Klasterisasi")
plt.show()
```
